```python
import numpy as np
```

```python
data=np.loadtxt("data.csv",delimiter=",")
```

```python
data.shape
```

```python
x=data[:,0].reshape(-1, 1)
y=data[:,1]
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn import model_selection
```

```python
X_train,X_test,Y_train,Y_test = model_selection.train_test_split(x, y)
#split data that is 75% is into train data and 25% is into test data
#this is a random split so each time the output is different because each time the train data and test data sperate randomply
```

```python
from sklearn.linear_model import LinearRegression
```

```python
alg1=LinearRegression()
```

```python
alg1.fit(X_train,Y_train)
```

```python
alg1.coef_
```

```python
alg1.intercept_
```

```python
import matplotlib.pyplot as plt
m=alg1.coef_
c=alg1.intercept_
x_line=np.arange(0,10,0.1)
y_line=m*x_line+c
plt.plot(x_line,y_line)
plt.show()
```
